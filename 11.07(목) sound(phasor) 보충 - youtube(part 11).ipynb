```python
t = np. arange(1, sr*dur+1)/sr # arange function은 가장 마지막을 배제시키기 때문에 +1이 필요
```

```python
theta = t * 2*np.pi * freq # 아까와 달리 t을 가지고 theta로 확장, freq가 없다고 생각하면 1초 동안에 한 바퀴 도는 것을 만들어라는 얘기 -> freq를 넣으면 바퀴에 해당되는 것을 곱하기
```

```python
s = np.sin(theta) # 아까와 달리 theta와 s에 더해 t까지 가진 상태, 
```

```python
# 지금까지는 sin, cos 공식을 이용한 phasor

# generate signal by complex-phasor
c = np.exp(theta*1j) #오일러 공식을 이용한 phasor #np.exp가 큰e라고 생각하기, 1j = i 
c # out의 값들은 a + bi의 형태 / -1이면 소숫점을 하나 더 왼쪽으로, +1이면 오른쪽으로 #정보량이 동일? # 복소수 complex number의 벡터가 들어가있
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.') # 3개의 입력이 들어감( , , ) <- 3차원 벡터 # 마찬가지로 size에서 corresponding
# c.real은 c에서 a만 빼오고(ex. 위에서 0.99802673) c.imag는 b만 빼옴(ex. 위에서 6.27905~) 
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```

```python
#sin과 달리 오일러 function은 하나의 각도값을 넣으면 a,b에 해당하는 2개가 나옴
```
